```python
import sqlite3
import pandas as pd 

connection = sqlite3.connect(':memory:')

cursor = connection.cursor()
```

```python
create_table_query = '''
CREATE TABLE Students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    STATISTICS_marks INTEGER,
    python_marks INTEGER,
    ml_marks INTEGER,
    dl_marks INTEGER
);
'''

cursor.execute(create_table_query)
print("Students table created.")
```

```python
insert_data_query= '''
INSERT INTO Students (name, age, statistics_marks, python_marks, ml_marks, dl_marks)
VALUES
('Mazen', 20, 85, 90, 88, 92),
('Khaled', 21, 75, 78, 82, 85),
('Hamza', 22, 88, 91, 84, 89),
('Achraf', 19, 92, 89, 95, 91),
('Abdo', 23, 80, 87, 85, 90);
'''

cursor.execute(insert_data_query)
connection.commit()

print("Data inserted into the Student table.")
```

```python
query = "SELECT * FROM Students;"
df = pd.read_sql(query, connection)
print("Querying the Students table:")
df
```

```python
update_query = "UPDATE Students SET python_marks = 95 WHERE name = 'Khaled'"
cursor.execute(update_query)
connection.commit()

print("Updated Khaled's Python grade")
```

```python
updated_df = pd.read_sql(query, connection)

print("Updated data")
updated_df
```

```python
cursor.close()
connection.close()
```
